```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/Users/msvar/Desktop/sample.csv')
print(dataset.head(10))
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset.LotArea, dataset.SalePrice, color='Blue', marker='o')
plt.show()
```

```python
X = dataset.LotArea.array.reshape(-1, 1)
print(X)
Y = dataset.SalePrice
print(Y)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
x = int(input("Enter the area of the house - "))
area = [[x]]
Price = model.predict(area)
print(Price)
```
